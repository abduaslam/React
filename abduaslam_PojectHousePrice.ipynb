```python
#1.6 splitting data to training and testing data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25) 

```

```python
#1.7 data scalling (make all data in small range,I used StandardScaler)

from sklearn.preprocessing import StandardScaler 

scl=StandardScaler()
scl.fit(x_train)
x_train=scl.transform(x_train)
x_test=scl.transform(x_test)
```

2.Model Selection 

```python
#2.1 selecting LinearRegression model(training data)
from sklearn.linear_model import LinearRegression 

model=LinearRegression()
model.fit(x_train,y_train)
```

3.Predect data (predicting t_test values)

```python
#3.1 predect model one
yp=model.predict(x_test)
```
